3. Place your data in a colab notebook into a pandas dataframe

```
import pandas as pd

url_of_AAPL_file = "https://raw.githubusercontent.com/lgelbman/DataScienceHW3/main/AAPL.csv"
data_AAPL = pd.read_csv(url_of_AAPL_file)

url_of_AMZN_file = "https://raw.githubusercontent.com/lgelbman/DataScienceHW3/main/AMZN.csv"
data_AMZN = pd.read_csv(url_of_AMZN_file)

url_of_FB_file = "https://raw.githubusercontent.com/lgelbman/DataScienceHW3/main/FB.csv"
data_FB = pd.read_csv(url_of_FB_file)

url_of_GOOG_file = "https://raw.githubusercontent.com/lgelbman/DataScienceHW3/main/GOOG.csv"
data_GOOG = pd.read_csv(url_of_GOOG_file)

url_of_MSFT_file = "https://raw.githubusercontent.com/lgelbman/DataScienceHW3/main/MSFT.csv"
data_MSFT = pd.read_csv(url_of_MSFT_file)

data_frames = [data_AAPL, data_AMZN, data_FB, data_GOOG, data_MSFT]
```

6. Use a linear regression to predict the output for the next five days of one of the stocks

```
data_AAPL
```

```
from sklearn.linear_model import LinearRegression
import numpy as np


#used index instead of date because linear regression requires numeric data
X = np.array(data_AAPL.index).reshape(-1, 1)

Y = np.array(data_AAPL[["Open"]])


model = LinearRegression().fit(X, Y)

last_day_in_data_set = X.max()
days_to_predict = range(last_day_in_data_set, last_day_in_data_set + 5)
for i in days_to_predict:
  i = np.array([[i]])
  y = model.predict(i)
  display(y)





```
